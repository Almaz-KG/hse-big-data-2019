### Sensors types and characteristics

`Sensor is device that measures or detect a property of the environment, or changes to a property`

Sensors
- Exteroceptive: (Extero means outside) = surroundings
- Proprioceptive: (Proprio means interal)

### Most common Exteroceptive sensors
- Camera
- LIDAR
- SONAR
- RADAR

#### Camera
- Passive light-collecting sensor
- Essential for correctly perceiving environment

Comparision metrics:
- Resolution - the quality of image
- Field of view (FoV) - horizontal and vertical angular extent that is visible to the camera
- Dynamic Range - is the difference between the lightest and the darkest tones in an image


Stereo camera
-  The combination of two cameras with overlapping fields of view and aligned image planes
- Allows to depth estimation from synchronized image pairs

Some groups belive that the camera is the only sensor needs for self-driving car's percention task.

#### LIDAR - Light Detection and Ranging Sensor

- Produces the detailed 3D scene geometry from cloud points
- LIDAR are not effected by the environment lighting

Comparision metrics:
- Number of beams (8, 16, 32, 64)
- Points per second
- Rotation rate
- Detection rate
- Field of View

#### Radar - Radio Detection and Ranging

- Robust Object Detection and Relative Speed Estimation
- Useful in adverse weather 

Comparision metrics:
- Detection Range
- Field of view
- Position and speed accuracy

Configurations:
- WFOV for short ranges
- NFOV for long ranges

#### Sonar - Sound Navigation and Ranging

- Measure range using sound waves
- Short-range all-weather distance measurement
- Ideal for low-cost parking solutions
- Unaffected by lighting, precipitation

Comparision metrics
- Detection range
- Field of view
- Costs

### Most common Proprioceptive sensors

- GNSS/IMU
- Wheel odometry

#### GNSS/IMU - Global Navigation Satellite Systems and Inertian Measurement Unit

- GPS, Galilie, GLONAS, etc
- GNSS receivers are used to measure ego vehicle 
 - position, 
 - velocity, 
 - heading
 - Varying accuracies: RTK, PPP, DGPS
- IMU are used to measure 
 - angular rotation rate
 - acceleration
 - heading

#### Wheel odometry sensors

- Tracks wheel velocities and orientation
- Estimating the speed and heading rate

### Self-driving computing hardware

#### Computing Hardware
- Need a `self-driving brain`
 - Takes all sensor data
 - Computes actions
 - Generates commands
- Already existing advanced systems that do self driving car processing 
 - Nvidia Drive PX/AGX
 - Intel & Mobileye EyeQ
 
- Computing tasks
 - Image processing
 - Object detection
 - Mapping
 
- Computing unit form-factors
 - GPUs - Graphic Processing Unit
 - FPGAs - Field Programmable Gate Array
 - ASICs - Application Specific Integrated Chip

#### Synchronization Hardware

- Synchronize different modules and provide a common clock
- Sensor measurements must be timestamped with consistent times for sensor fusion to function correctly
