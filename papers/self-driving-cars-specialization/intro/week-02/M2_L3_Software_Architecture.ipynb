#### Environment Perception module
- Localization
- Dynamic Object Detection
- Static Object Detection
- Dynamic Object Tracking 
- Object Motion Prediction 

![Environment Perception](resources/images/environment-perception.png)

#### Environment Mapping module

- creates a set of maps which locate objects in scene 

- Occupancy Grid Map - static objects in the environment
- Localization Map 
- Detailed Road Map

![Environment Mapping](resources/images/environment-mapping.png)

#### Motion Planning module

- By the information from Perception and Mapping modules decides where to drive

- Mission planner - Long-term planning
- Behavior Planner - Short-term planning (merge, switching to other lane, etc)
- Local Planner -Reactive planning (smooth, safe, efficient maneuver). The output of this part is a planned trajectory which is a combined desired path and velocity profile for a short period of time into the future. 

![Motion Planning Module](resources/images/motion-planning.png)

#### System Supervisor

- Continuously monitors all aspects of car
- Gives the appropriate warning in the event of subsystem failure

Two Parts
- Hardware supervisor
 - Monitors sensors for any faults
   - broken sensor,
   - missing measurement
   - degraded information
 - Analyze the hardware outputs
- Software supervisor
 - Responsible for validating the software stack
 - Analyze the inconsistencies between the outputs of all modules
