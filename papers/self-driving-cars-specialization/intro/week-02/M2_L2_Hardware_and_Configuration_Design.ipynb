#### Driving assumptions

- Aggressive deceleration = $5 \frac{m}{s^2}$
- Comfortable deceleration = $2\frac{m}{s^2}$

- Stopping distance: $d = \frac{v^2}{2a} \;\;\;\; v\text{ - is velocity, a - deceleration rate}$
