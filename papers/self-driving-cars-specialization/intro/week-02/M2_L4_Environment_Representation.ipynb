### Environment Representation

#### Environmental Map Types

- Localization of vehicle in the environment
 - Localization point cloud or feature map

![localization map](resources/images/localization_map.png)

- Collision avoidance with static objects
 - Occupancy Grid Map is a 2D or 3D discretized map of the static objects in the environment of self driving car. 

![localization map](resources/images/accupancy_grid.png)

- Path planning
 - Detailed Road Map

The Detailed Road Map might be created in one of three ways
- Fully Offline
- Fully Online
- Created offline and Upadated Online
