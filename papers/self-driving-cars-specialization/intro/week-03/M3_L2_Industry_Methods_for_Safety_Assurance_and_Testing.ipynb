### Module 3: Safety Assurance for Autonomous Vehicle

### Industry perspectives on self driving safety

Waymo Safety Levels

- Behavioral Safety
- Functional Safety
- Crash Safety
- Operational Safety
- Non collision Safety

Waymo Safety Processes
- Identify hazard scenarious and potential mitigations
- Use hazad assesment methods to define safety requirements
 - Preliminary analysis
 - Fault tree
 - Design Failure Modes and Effects Analyses
- Conduct extensive testing to make sure safety requirements are met

Waymo: Levels of testing to ensure safety
- Simulation testing
- Closed-course testing
 - Follow 28 core + 19 additional scenario
 - Focus on four most common crashes: 
   - Rear-end 
   - Intersection
   - Road departure
   - Lane change
- Real-world driving
 - Start with small fleet, expand steadily
 - Already testing thousands of vehicles

GM Safery
- Address all 12 elements of NHTSA Safety Framework
- Iterative Design
 - 1. Analyse
 - 2. Build
 - 3. Simulate
 - 4. Drive
 - 5. Go p.1
- Control car production
- Safety through Comprehensive Risk Management and Deep Integration
 - identify and address risks, validate solutions
 - prioritize elimination of risk, not just mitigation
- All hardware, software systems meet
 - self-set standards for performance, crash protection, reliability, serviceability, security, safety

GM Safety Processes
- Deductive Analysis
 - Fault Tree analysis
- Inductive analysis
 - Design and process FMEA
- Exploratory Analysis
 - HAZOP: Hazard and Operability Study

GM: Safety Thresholds

All GM vehicels are equipped with two key safety thresholds
 - Fail safes (redundant functionality)
 - SOTIF - all critical functionalities are evaluated for unpredictable scenarios

GM: Testing process
- Performance testing
- Requirements validation of components
- Fault injection testing of safety critical functionality
- Intrusive testing such as electromagnetic interferenece, etc
- Durability testing 
- Simutation based testing

### Approaches to demonstrating autonomy safety

**Analytical Safety** - Ensuring the system works in theory and meets safety requirements found by hazard assessment

**Data Driven Safety** - Safety guarantee due to the fact that the system has performed autonomously without fail on the roads for a very large number of kms

Are autonomous cars safer?

- Driving is still dangerous
- Car accidents are mostly (90%) caused due to human errors
- In US, on average (human driving statistics)
 - 1 fatal collision per 146 million km
 - 1 injury collision per 2.1 million km
 - ~1 collision per 400 000 km
- Based on statistics published by all autonomous driving vehicles beeing tested in California: 
 - In, 2017, Waymo had
   - Driven 563 000 km autonomouly
   - 63 disengagements
   - 1 disengagement every 9000 km
 - In, 2017, GM had
   - Driven 210 000 km autonomouly
   - 105 disengagements
   - 1 disengagement every 2000 km
