**FMEA - Failure Mode and Effects Analyses**

- Bottom up process to identify all the effects of faults in a system
- Failure mode - modes or ways in which a component of the system may fail
- Effect Analysis - analysing effects of the fauilure modes on the operation of the system

FMEA Idea

- Categorize failure modes by priority
 - How serious are their effects?
 - How frequently do they happen?
 - How easily can they be detected?
- Eliminate or reduce failures, starting with top priority

FMEA Steps:
- Discuss with field experts
- Create FMEA table
- Make a List of failure modes
- For each failure mode
 - Identify effects
 - Which type of severity(S) those effects?
 - For each root cause
   - What is the occurrence (O)?
   - For each root cause
     - What is the detection (D) method?
- Compute the Risk Priority Number (RPN) = S * O * D
     

**The example of FMEA**

- Considering following failure mode
 - System encounters gravel, controller failure
   - Severity: Physical crash => S = 10
   - Occurence: Whenever construction encountered, out of ODD, so 0 ~= 4
   - Detection: Can check status monitor to identify if this happens with D = 10
 - Risk priority number (RPN) = S * O * D = 10 * 4 * 10 = 400
- Sign perception failure
  - RPN ~= 100
- GPS synchronization failure
  - RPN ~= 300
- Incorrect motion prediction 
  - RPN ~= 150
- Pedestrian prediction failure
  - RPN ~= 900
  
The final RPN list:
- Pedestrian prediction failure
- Control failure
- GPS synchronization failure
- Incorrect motion prediction 
- Sign perception failure


**HAZOP - Hazard and Operability Study**

- Is a variation on FMEA
- Qualitative brainstorming process, needs "imagination"
- Uses guide words to trigger brainstorming (not, more, less, etc)
- Applied to complex 'processes'
 - Sufficient design information is avialable, and not likely to change significantly

### Autonomous/Automotive Safety Frameworks

- ISO 26262 - Functional Safety Standard
- ISO/PAR 21448.1 - Safety of Intended Functionality (SOTIF)

**Functional Safety** is defined as:
- safety due to absense of unreasonable risk
- only concerned about malfunctioning system

ISO 26262 defines four Automotive Safety Integrity Levels (ASIL)
- ASIL-D most stringent
- ASIL-A least stringent

**Functional Safety process**

![Functional Safety Process](resources/images/functional-safety-process.png)

The **Summary Functional Safety Assessment** - process to evaluate residual risk and determine if system has reached an acceptable level of safety

The **Hazard and Risk Assessment(HARA)** - process for identifying and categorizing hazardous events and specify requirements to avoid unreasonable risk

**Safety of the Intended Functionality (SOTIF)**

- SOTIF is specifically concenrned with failure causes related to system performance limitations and prefictable misuse of the system
 - sensor limitations 
 - algorithm failures/insufficiencies
 - user misuse - overload, confusion
 
- Designed for level 0-2 autonomy
- Extension of FUSA
 - V-shaped process
 - Employs HARA
