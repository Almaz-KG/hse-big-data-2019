### Formal definitions

- Harm - the physical harm to a living thing
- Risk - the probability that an event occurs, combined with the severity of the harm that the event can cause
- Safety - absence of unreasonable risk of harm
- Hazard - potential source of unreasonable risk of harm

### Major hazard sources

- Mechanical
- Electrical
- Hardware
- Software
- Sensors
- Behavioral (planning, decision-making)
- Fallback
- Cyber

### Safety requirements

NHTSA - National Highway Transportation Safety Administration

NHTSA has defined a 12 part safety framework
- System Design/Engineering approach
- Autonomy Design
 - ODD
 - OEDR
 - Fallback
 - Traffic Laws
 - Cybersecuruty
 - HMI (Human Machine Interface)
- Testing & Crash Mitigation
 - Testing (simulation, close track testing, public road driving)
 - Crashworthiness
 - Post crash behaviour
 - Data recording - black box
 - Consumer Education and Training
