#### OEDR - is a criterion for automation

What is perception?
- We want to make sense of the environment and ourselves
- That leads us into two things:
 - object identification (what is it: car, truck, bike, etc)
 - object motion (where is moving this object?)

Goals for perception

- Perception of static objects
 - Road and lane markings (on-road)
 - Curbs (off-road)
 - Traffic lights (off-road)
 - Road signs (off-road)
 - Construction signs, obstructions, and more (on-road)
 
- Perceptions of dynamic objects (on-road)
 - Other vehicles (4 wheelers, 2 wheelers)
 - Pedestrians

- Ego localization
 - Ego Position
 - Velocity, acceleration
 - Orientation, angular motion
 

#### Challenges to perception

- Robust detection and segmentation
- Sensor uncertainty
- Occlusion, reflection on camera and other sensors
- Illuminations, lens flare
- Weather, precipitation
