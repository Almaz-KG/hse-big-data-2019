## Course 1: Introduction to Self-Driving Cars

### Module 1: Driving Taxonomy, Perception and Driving Decisions 

#### Making decisions
- Long term 
 - How to navigate from NY to Los Angeles?
- Short term
 - Is it safe to turn left now?
 - Can I pass this intersection and join the left road?
- Immediate desicions
 - Can I stay on track on this curved road?
 - Accelerate or brake, by how much?

#### There are many rules to make a right driving decision

- safety
- efficiently
- following all listed traffic rules
- only follow those rules everyone else is following

**Driving decision making is complicated**

#### Reactive Planning
- Decision Tree Based Planning
- Rule based planning

Examples:
- If there is a pedestrian on the road - STOP!
- If speed limit changes - adjust speed to match it!

#### Predictive Planning

- Make prediction about other vehicles and how they are moving. Then use these predictions to inform our decisions

Examples:
- That car has been stopped for the last 10 seconds. It is going to bo stopped for the next few seconds.
- This pedestrian is jaywalking. He will enter our lane by the time we reach him.
