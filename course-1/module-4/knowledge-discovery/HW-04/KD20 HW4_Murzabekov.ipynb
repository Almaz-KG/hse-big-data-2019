### Optimal Estimation

Consider the following series of measurements of the unknown value $x$: 

$$y_i =a_ix+ε_i, \;\;\;\;i=1,...,n,$$
where $y_i$ are measurement results, $a_i$ are known coefficients, and $ε_i$ represent random error of measurement and are independent identically distributed (i.i.d.) with zero mean
and variance $σ^2$:

$$Eε_i =0, Eε^2_i =σ^2, \;\;\;\;i=1,...,n$$

**a)** What function of $y_1,...,y_n$ (and of $a_1,...,a_n$) would you use as a good estimate $\widehat{x}$
for $x$? $\widehat{x}$ = ? 


(Hint: try best linear estimate or least squares estimate).


Estimation of $x$ might be formulated in the following formula, and the value of this formula we should minimize

$$
\begin{equation*}
Q(x) = \sum_{i=1}^n \frac{(y_i - a_ix) ^ 2}{σ^2} \to min
\end{equation*}
$$

$$
\frac{dQ}{dx} = \sum_{i=1}^n\frac{(-2y_ia_i + 2a^2_ix)}{σ^2} = \sum_{i=1}^n\frac{-2(y_ia_i - a^2_ix)}{σ^2}=-2\sum_{i=1}^n\frac{y_ia_i - a^2_ix}{σ^2} = 0 \Rightarrow
$$

$$
-2\sum_{i=1}^n\frac{y_ia_i - a^2_ix}{σ^2} = 0 \Rightarrow
$$

$$
-2(\sum_{i=1}^n\frac{y_ia_i}{\sigma^2} - \sum_{i=1}^n\frac{a^2_ix}{\sigma^2}) = 0 \Rightarrow
$$

$$
-2 \neq 0 \Rightarrow
(\sum_{i=1}^n\frac{y_ia_i}{\sigma^2} - \sum_{i=1}^n\frac{a^2_ix}{\sigma^2}) = 0 \Rightarrow
$$

$$
(\sum_{i=1}^n\frac{y_ia_i}{\sigma^2} - \sum_{i=1}^n\frac{a^2_ix}{\sigma^2}) = 0 \Rightarrow
$$

$$
\sum_{i=1}^n\frac{y_ia_i}{\sigma^2} = \sum_{i=1}^n\frac{a^2_ix}{\sigma^2}\Rightarrow (* \sigma^2)
$$

$$
\sum_{i=1}^n{y_ia_i} = \sum_{i=1}^n{a^2_ix}\Rightarrow
$$

$$
\sum_{i=1}^n{y_ia_i} = x\sum_{i=1}^n{a^2_i}\Rightarrow
$$

$$ x\sum_{i=1}^n{a^2_i} = \sum_{i=1}^n{y_ia_i}\Rightarrow $$

$$ 
\widehat{x} = \frac{\sum_{i=1}^n{y_ia_i}}{\sum_{i=1}^n{a^2_i}}
$$

**b)** Is this estimate optimal in any sense?

In this work I'm using the method **Least squares estimate**. This method/function approximation applies the principle of least squares to function approximation, by means of a weighted sum of other functions. The best approximation can be defined as that which minimises the difference between the original function and the approximation; for a least-squares approach the quality of the approximation is measured in terms of the squared differences between the two.
