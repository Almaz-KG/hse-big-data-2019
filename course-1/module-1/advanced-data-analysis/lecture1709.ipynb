```python
import os
import json
from functools import reduce
from collections import Counter
```

```python
path = os.getcwd()
path = os.chdir("C:/Users/student/Desktop/Gorbunov/data")
```

```python
tweets = []
with open("winter_olympics.json", "rb+") as readfile:
    #res = readfile.readlines()
    i = 0 
    for line in readfile:
        tweets.append(json.loads(line.decode("utf-8")))
        i+=1
        if i==5000:
            break
        
```

```python
timestamp = list(map(lambda x: x.get("created_at", "") , tweets[:10000]))
```

```python
retweet_count = list(map(lambda x: x.get("retweet_count", "") , tweets[:10000]))
```

```python
reduce(lambda x,y: x+y, retweet_count)/len(tweets)
```

```python
entities = list(map(lambda x: x.get("entities", "") , tweets[:10000]))
```

```python
hashtags = list(map(lambda x: [hashtag.get("text","").lower() for hashtag in x.get("entities", "").get("hashtags", "")] , tweets[:10000]))
```

```python
hashtags
```

```python
user_mentions = list(map(lambda x: [hashtag.get("name","").lower() for hashtag in x.get("entities", "").get("user_mentions", "")] , tweets[:10000]))
```

```python
user_mentions_count = Counter()
[user_mentions_count.update(param) for param in user_mentions]
```

```python
hashtags_count = Counter()
[hashtags_count.update(param) for param in hashtags]
```

```python
filter_hastags = set(map(lambda x: x[0] ,hashtags_count.most_common(20)))

list(filter(lambda x: set(x[1]).issubset(filter_hastags), list(zip(timestamp, hashtags))))

list(filter(lambda x: len(set(x[1]).intersection(filter_hastags))!=0 , list(zip(timestamp, hashtags))))
```
