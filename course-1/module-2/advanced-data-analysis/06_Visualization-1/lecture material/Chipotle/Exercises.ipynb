# Visualizing Chipotle's Data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import collections
import matplotlib.pyplot as plt 

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this data folder chipotle.tsv

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(".//data/chipotle.tsv",  sep="\t")
chipo
```

### Step 5. Create a histogram of the top 5 items bought

```python
x = chipo.item_name
letter_counts = collections.Counter(x)
letter_counts
```

```python
df = pd.DataFrame.from_dict(letter_counts, orient="index") 
```

```python
df = df[0].sort_values(ascending=False).head(5)
```

```python
df.plot(kind="bar")

plt.xlabel("Items")
plt.ylabel("Price")
plt.title("Most ordered Chipolte\'s Items")
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo.item_price = [float(value[1:]) for value in chipo.item_price] 
```

```python
chipo.info()
```

```python
orders = chipo.groupby("order_id").sum()
```

```python
plt.scatter(x=orders.item_price, y=orders.quantity, s=100, c="green")
plt.xlabel('Order Price')
plt.ylabel("Items ordered")
plt.title('Number of items ordered per order price')
```
