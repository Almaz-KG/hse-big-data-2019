# Getting and Knowing your Data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this data folder chipotle.tsv

```python
data = pd.read_csv("data/chipotle.tsv", sep = "\t")
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = data
```

### Step 9. Which was the most ordered item?

```python
c = chipo.groupby("item_name").sum()
c = c.sort_values(["quantity"], ascending = False)

c.head(10)
```

### Step 11. What was the most ordered item in the choice_description column?

```python
cd = chipo.groupby("choice_description").sum()
cd = cd.sort_values(["quantity"], ascending = False)
cd.head(1)
```

### Step 12. How many items were orderd in total?

```python
chipo[["quantity"]].sum()

total_items_ordered = chipo["quantity"].sum()
```

### Step 13. Turn the item price into a float

```python
chipo["item_price"] = chipo["item_price"].apply(lambda x: float(x[1:]))
```

```python
chipo.head(10)
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue_per_row = chipo.item_price * chipo.quantity

revenue = revenue_per_row.sum()

int(revenue)
```

### Step 15. How many orders were made in the period?

```python
orders = chipo.order_id.value_counts().size
orders
```

### Step 16. What is the average amount per order?

```python
order_grouped = chipo.groupby(["order_id"]).sum()

np.round(order_grouped.mean()["quantity"], 2)
```
