# Filtering and Sorting Data

### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this data folder chipotle.tsv 

```python
data = pd.read_csv("data/chipotle.tsv", sep = "\t")
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = data
```

```python
chipo.head(3)
```

### Step 4. How many products cost more than $10.00?

```python
prices = [float(value[1:]) for value in chipo.item_price]

chipo.item_price = prices
```

```python
chipo.head(3)
```

```python
chipo10 = chipo[chipo.item_price > 10.0]

chipo10.head(3)
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_filtered = chipo10.drop_duplicates(["item_name","item_price"])

chipo_filtered.head(3)
```

```python
chipo_one = chipo_filtered[chipo_filtered.quantity == 1]

price_per_item = chipo_one[["item_name","item_price"]]

price_per_item.sort_values("item_price", ascending = False)

price_per_item.head(3)
```

### Step 6. Sort by the name of the item

```python
chipo.item_name.sort_values().value_counts()
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values("item_price", ascending = False).head(1)
```
